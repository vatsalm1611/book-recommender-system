```python
#only selecting ratings from users who have given more than 200 ratings
ratings = ratings[ratings['user_id'].isin(y)]
```

```python
#now join ratings with books
ratings_with_books = ratings.merge(books, on='ISBN')
```

```python
ratings_with_books
```

```python
#so we are basically making a single dataframe with three dataframes
#that's why we are doing filtering, merging etc
```

```python
number_rating = ratings_with_books.groupby('title')['rating'].count().reset_index()
#this will give number of ratings for each book
```

```python
number_rating.head()
#so this will give number of ratings for each book
```

```python
number_rating.rename(columns = {'rating': 'num_of_ratings'}, inplace = True)
#this is done to avoid confusion
```

```python
final_rating = ratings_with_books.merge(number_rating, on='title')
#this is the final rating dataframe which we will use for recommendation
```

```python
final_rating.head()
```

```python
#now i only need those books which have atleast 50 ratings
final_rating = final_rating[final_rating['num_of_ratings'] >= 50]
```

```python
# Filter top 1000 users and top 1000 books by rating counts from final_rating
top_users = final_rating['user_id'].value_counts().head(1000).index
top_books = final_rating['title'].value_counts().head(1000).index

final_rating_filtered = final_rating[final_rating['user_id'].isin(top_users) & final_rating['title'].isin(top_books)]

# Ab book_pivot final_rating_filtered se banao:
# book_pivot = final_rating_filtered.pivot_table(columns='user_id', index='title', values='rating')
```

```python
#pivot table -> 1. it is like a multidimensional table which helps to summarize data and make comparisons for easier analysis
#2. it is used to change the data format
#3. it is used to aggregate data
#4. it is used to group data
#5. it is used to filter data
#etc
#in our project pivot table is used to create a matrix of user_id and title with rating as values
#which will help us to find similar books based on user ratings
#so ratings is basically our root point for recommendation system
```

```python
#now let's build the recommendation system using this pivot table
```

```python
try:
    book_pivot = final_rating_filtered.pivot_table(columns='user_id', index='title', values='rating')
    print('book_pivot created successfully!')
    print('Shape:', book_pivot.shape)
except Exception as e:
    print('Error creating book_pivot:', e)
    print('final_rating_filtered shape:', final_rating_filtered.shape)
    print('Columns:', final_rating_filtered.columns.tolist())
    print('Sample data:')
    display(final_rating_filtered.head())
```

```python
book_pivot
```

```python
#NaN data can not be passed to the model so we wil convert it to 0
```

```python
book_pivot.shape
```

```python
book_pivot.fillna(0, inplace = True)
#we are filling NaN values with 0 because we want to treat books that a user has not rated as 0 rating
```

```python
book_pivot
```

```python
#now i will start training the model
#but before that i will have to convert the pivot table to CSR format
```

```python
from scipy.sparse import csr_matrix
```

```python
book_sparse = csr_matrix(book_pivot)
```

```python
book_sparse
```

```python
type(book_sparse)
```

```python
from sklearn.neighbors import NearestNeighbors
model = NearestNeighbors(algorithm = 'brute')
#we are using brute force algorithm because it is faster than other algorithms for small datasets
#it is also more accurate than other algorithms for small datasets
```

```python
model.fit(book_sparse)
#fit the model with the sparse matrix
#so now the model is trained and we can use it for recommendation
```
