```python
try:
    book_pivot = final_rating_filtered.pivot_table(columns='user_id', index='title', values='rating')
    print('book_pivot created successfully!')
    print('Shape:', book_pivot.shape)
except Exception as e:
    print('Error creating book_pivot:', e)
    print('final_rating_filtered shape:', final_rating_filtered.shape)
    print('Columns:', final_rating_filtered.columns.tolist())
    print('Sample data:')
    display(final_rating_filtered.head())
```

```python
book_pivot.fillna(0, inplace = True)
#we are filling NaN values with 0 because we want to treat books that a user has not rated as 0 rating
```

```python
#now i will start training the model
#but before that i will have to convert the pivot table to CSR format
```

```python
from scipy.sparse import csr_matrix
```

```python
book_sparse = csr_matrix(book_pivot)
```

```python
book_sparse
```

```python
type(book_sparse)
```

```python
from sklearn.neighbors import NearestNeighbors
model = NearestNeighbors(algorithm = 'brute', metric='cosine')
#we are using brute force algorithm with cosine similarity for better recommendations
```

```python
model.fit(book_sparse)
#fit the model with the sparse matrix
#so now the model is trained and we can use it for recommendation
```

```python
book_pivot.iloc[237,:]
#this will give the ratings of the book at index 237
#now we will use this data to find similar books
```

```python
distances, suggestions = model.kneighbors(book_pivot.iloc[237,:].values.reshape(1,-1), n_neighbors = 6)
#we are reshaping the data because the model expects a 2D array
#n_neighbors is the number of similar books we want to find
```

```python
def recommend_book(book_name):
    try:
        # Find the index of the book in pivot table
        book_id = np.where(book_pivot.index == book_name)[0][0]
    except IndexError:
        print(f"Book '{book_name}' not found in the data.")
        return

    # Get distances and suggestions using trained model (n_neighbors=10 for better results)
    distance, suggestion = model.kneighbors(
        book_pivot.iloc[book_id, :].values.reshape(1, -1),
        n_neighbors=10
    )

    display_count = 6  # Only display top 5-6 recommendations (excluding the searched book itself)
    shown = 0
    for i in range(len(suggestion)):
        books = book_pivot.index[suggestion[i]]
        for j in books:
            if j == book_name:
                print(f"You searched '{book_name}'\n")
                print("The suggestion books are:\n")
            else:
                if shown < display_count:
                    print(j)
                    shown += 1
```

```python
book_name = "1984"
recommend_book(book_name)
#this will give the recommendations for the book 1984
```
